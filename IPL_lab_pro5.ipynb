# **Data Cleaning & Preprocessing**

```python
# Data Handling
import pandas as pd
import numpy as np

# Date and Time Manipulation
from datetime import datetime

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8')  # cleaner look

# Interactive Visuals (optional, if needed)
import plotly.express as px

# Display settings
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

# Statistical analysis
import scipy.stats as stats
```

```python
# Load the dataset
matches=pd.read_csv('/content/matches (1) - matches (1).csv.csv')
# Display
matches
```

```python
#load the dataset
deliveries=pd.read_csv('/content/deliveries - deliveries.csv.csv')

# Display
deliveries
```

```python
# Convert Date to datetime format
matches['date'] = pd.to_datetime(matches['date'], errors='coerce')

# Format as dd/mm/yyyy
matches['date'] = matches['date'].dt.strftime('%d/%m/%Y')

# Check first few rows
print(matches[['date']].head())

matches.info()
```

```python
#check for null values
matches.isnull().sum()
```

```python
#Removing unwanted columns
matches.drop(['umpire3'], axis=1, inplace=True)# No entries registered
```

```python
#Replacing Null values
matches['city'].fillna('UAE', inplace=True)#since venue is same additionally dates are consecutive and happeining in same season - filling with exact city name from outside source.
matches['umpire1'].fillna('Virender Sharma', inplace=True)   # from outside source
matches['umpire2'].fillna('Sundaram Ravi', inplace=True)     # from outside source
matches['winner'].fillna('No result', inplace=True)          # since match cancelled
matches['player_of_match'].fillna('No result', inplace=True) # since match cancelled
```

```python
#Recheck for null values again
matches.isnull().sum()
```

```python
#checking for duplicates
matches.duplicated().sum()
```

```python
#check for null values
deliveries.isnull().sum()
```

```python
#Replacing Null values
deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna("No dismissal")  #blanks indicates that no dismissal happened in that ball.
deliveries['dismissal_kind']  = deliveries['dismissal_kind'].fillna("No dismissal")     #blanks indicates that no dismissal happened in that ball.
deliveries['fielder'] = deliveries['fielder'].fillna("No fielder")                      #No fielder involved in that dismissal (dismissals like caught and bowled,bowled,hit wicket,LBW)
```

```python
#Recheck for null values again
deliveries.isnull().sum()
```

```python
#checking for duplicates
deliveries.duplicated().sum()
```

```python
# Show the duplicated row(s)
deliveries[deliveries.duplicated()]
```

```python
# Show both the original and its duplicate
deliveries[deliveries.duplicated(keep=False)]
```

```python
#clearing data redundancy
deliveries = deliveries.drop_duplicates()
```

```python
#checking for duplicates again
deliveries.duplicated().sum()
```

```python
# Define replacement mapping
team_name_corrections = {
    "Rising Pune Supergiants": "Rising Pune Supergiant",
    "Rising Pune Supergaints": "Rising Pune Supergiant"   # in case of typo
}

# Apply correction in matches dataset
matches.replace(team_name_corrections, inplace=True)

# Apply correction in deliveries dataset
deliveries.replace(team_name_corrections, inplace=True)
```

**Player Selection & Auction Strategies**

```python
# All-rounders (players with both 500+ runs & 20+ wickets)
runs = deliveries.groupby('batsman')['batsman_runs'].sum()
wickets = dismissals.groupby('bowler').size()
all_rounders = pd.DataFrame({'runs': runs, 'wickets': wickets}).fillna(0)
all_rounders = all_rounders[(all_rounders['runs'] > 500) & (all_rounders['wickets'] > 20)]
print(all_rounders.sort_values(['runs','wickets'], ascending=False))
```

```python
runs = deliveries.groupby('batsman')['batsman_runs'].sum()
wickets = dismissals.groupby('bowler').size()
all_rounders = pd.DataFrame({'runs': runs, 'wickets': wickets}).fillna(0)
all_rounders = all_rounders[(all_rounders['runs']>500) & (all_rounders['wickets']>20)]

plt.figure(figsize=(8,6))
sns.scatterplot(data=all_rounders, x="runs", y="wickets", size="runs", hue="wickets", palette="rainbow", legend=False)
plt.title("All-Rounders (Runs vs Wickets)")
plt.xlabel("Runs Scored")
plt.ylabel("Wickets Taken")
plt.show()
```

Highlights high-value auction players (strong in both batting & bowling).

**Performance Analysis & Optimization**

```python
# Strike rate by batsman
strike_rates = deliveries.groupby('batsman').apply(
    lambda x: (x['batsman_runs'].sum() / max(1, x.shape[0])) * 100
).sort_values(ascending=False).head(10)
print(strike_rates)
```

```python
strike_rates = deliveries.groupby('batsman').apply(
    lambda x: (x['batsman_runs'].sum() / max(1, x.shape[0]))*100
).sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=strike_rates.values, y=strike_rates.index, palette="crest")
plt.title("Top 10 Batsmen by Strike Rate")
plt.xlabel("Strike Rate")
plt.show()
```

High strike rate = impactful T20 batsman.
Helps optimize batting order.

**Augmented Coaching (Data-Driven Analysis)**

```python
# Average runs in powerplay (overs 1â€“6)
powerplay = deliveries[(deliveries['over'] >= 1) & (deliveries['over'] <= 6)]
pp_runs = powerplay.groupby('match_id')['total_runs'].sum().mean()
print("Average Powerplay runs:", pp_runs)

# Death overs (overs 16â€“20)
death = deliveries[(deliveries['over'] >= 16) & (deliveries['over'] <= 20)]
death_runs = death.groupby('match_id')['total_runs'].sum().mean()
print("Average Death overs runs:", death_runs)
```

```python
# Powerplay (1-6 overs) vs Death overs (16-20 overs)
powerplay = deliveries[(deliveries['over']>=1) & (deliveries['over']<=6)]
pp_runs = powerplay.groupby('match_id')['total_runs'].sum()

death = deliveries[(deliveries['over']>=16) & (deliveries['over']<=20)]
death_runs = death.groupby('match_id')['total_runs'].sum()

# Combine into DataFrame
pp_vs_death = pd.DataFrame({
    "Powerplay (1-6)": pp_runs,
    "Death Overs (16-20)": death_runs
})

# Melt for seaborn
pp_vs_death_melt = pp_vs_death.melt(var_name="Phase", value_name="Runs")

# Violin plot
plt.figure(figsize=(8,6))
sns.violinplot(x="Phase", y="Runs", data=pp_vs_death_melt, palette="muted")
plt.title("Distribution of Runs: Powerplay vs Death Overs")
plt.ylabel("Runs per Match")
plt.show()
```

Powerplay & death overs are critical phases in T20.
Teams can optimize batting/bowling strategy based on these stats.
